```python
#Import College csv file into the College MySQL table.

import mysql.connector as mysql
import csv
import pymysql

mydb = mysql.connect(
    host = "localhost",
    user = "root",
    password = "put your password",
    database = "academics"
)
cursor = mydb.cursor()


mycursor = csv.reader(open("C:\\Users\Emily Kimani\Desktop\Adv_Python\Mysql\College.csv"))

next(mycursor)
for row in mycursor:
    
    cursor.execute('INSERT INTO college(College_Id,College_Name,College_Address,College_City,College_Country) VALUES(%s,%s,%s,%s,%s)',row)

#close the connect to the database.
mydb.commit()
cursor.close()
print("Success")
```

```python
#Perform a 3 Table JOIN using the COLLEGE_ID field to join the College, Professor1 and Students MySQL tables, then display the records on the screen

import mysql.connector as mysql


mydb = mysql.connect(
host = "localhost",
user = "root",
password = "put your password",
database = "academics"

)

mycursor = mydb.cursor()

mycursor.execute("""SELECT *
    FROM (college
    INNER JOIN professor1 
        ON college.college_id = professor1.college_id)
    INNER JOIN students
        ON professor1.college_id = students.college_id""");

print(mycursor.fetchall())

mydb.commit()   
mycursor.close()

```
